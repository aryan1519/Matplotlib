Q5.)Laptop Sales

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(["ggplot"])
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]

company = np.array(company)
sold = np.array(sold)

plt.pie(sold,labels = company,autopct = "%.1f")

percentages = np.true_divide(sold,sold.sum(axis = 0,keepdims = True))*100 

for i in range(len(company)):
    print(company[i],'%.1f'%percentages[i])
plt.show()

```

Q7.)Plot Job_Year

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("Dataset/amazon_jobs_dataset.csv")

Years = data["Posting_date"]

years = Years.str.split(',',expand = True)[1]
freq = years.value_counts(ascending = True)

freq.sort_index(inplace = True)

x = freq.index
y = freq

plt.plot(x,y)
plt.title("Year vs job openings")
plt.xlabel("Years")
plt.ylabel("No of job openings")

plt.show()

for i in range(len(x)):
    print(x[i],y[i])
```

Q8.)Plot Month_Job

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(["ggplot"])

data = pd.read_csv("Dataset/amazon_jobs_dataset.csv")

Months = data["Posting_date"]

months = Months.str.split(expand = True)[0]
freq = months.value_counts()
x = freq.index
y = freq.values

plt.bar(x,y)
plt.xticks(rotation = "vertical")
plt.show()

for i in range(len(x)):
    print(x[i],y[i])
```

Q9.)Plot city_job

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(["ggplot"])

data = pd.read_csv("Dataset/amazon_jobs_dataset.csv")

loc = data["location"].str.split(", ",expand = True)

arr = loc[0] == "IN" # Boolean array

cities = loc[arr][2]

freq = cities.value_counts()

vals = freq.values
labels = freq.index
plt.pie(vals,labels = labels,autopct = "%.2f")
plt.show()

total = vals.sum()

for i in range(len(vals)):
    per = '%.2f'%((vals[i]/total)*100)
    print(labels[i],per)


```

Q10.)Plot Job_Java

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("Dataset/amazon_jobs_dataset.csv")

data["BASIC QUALIFICATIONS"].fillna("",inplace = True)

data = data[(data["BASIC QUALIFICATIONS"].str.contains("Java")) | (data["BASIC QUALIFICATIONS"].str.contains("java"))]
    
Years = data["Posting_date"]

years = Years.str.split(',',expand = True)[1]
freq = years.value_counts(ascending = True)

freq.sort_index(inplace = True)

yrs = freq.index
y = freq

plt.scatter(yrs,y)
plt.grid()
plt.show()
for i in range(len(yrs)):
    print(yrs[i],y[i],end =" ")
    print()
```
